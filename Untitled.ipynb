```python
import manage
import numpy as np

sar_img = manage.get_chandrayaan2Obj().date_map['20210312']
sar_mul = sar_img.multilook_azimuth(nmls=20, method='mean')

sar_mul_hh_mag = np.abs(sar_mul.HH)
```

```python
sar_mul_hh_mag.shape, sar_mul_hh_mag.dtype
```

```python
import matplotlib.pyplot as plt

dir = manage.get_dirs()
plt.imsave(dir.data.base / "image_data" / "sar_mul_hh_mag.png", sar_mul_hh_mag, cmap='tab20c')
```

```python
plt.imsave(dir.data.base / "image_data" / "sar_mul_hh_mag_10.png", sar_mul_hh_mag[:1645, :], cmap='tab20c')
```

```python
sar_mul_hh_mag.mean(), sar_mul_hh_mag.min(), sar_mul_hh_mag.max(), sar_mul_hh_mag.std()
```

```python
plt.hist(sar_mul_hh_mag.ravel(), bins=10000, log=True, fc='k', ec='k')
```

```python
plt.hist(sar_mul_hh_mag[:1645, :].ravel(), bins=10000, log=True, fc='k', ec='k')
```

```python
changed = sar_mul_hh_mag[:600, :].copy()
threshold = 20000
changed = np.where(changed < threshold, changed, threshold)
plt.imsave(dir.data.base / "image_data" / "experiment.png", changed, cmap='tab20c')
plt.imshow(changed, cmap='tab20c')
```

```python
# log 
sarnew = sar_mul_hh_mag.copy() + 1
sarnew = np.log(sarnew)
plt.imsave(dir.data.base / "image_data" / "log.png", sarnew, cmap='gnuplot2')
plt.imshow(sarnew, cmap='tab20c')
```

```python
# log10
sarnew = sar_mul_hh_mag.copy() + 1
sarnew = np.log10(sarnew)
plt.imsave(dir.data.base / "image_data" / "log10.png", sarnew, cmap='gray')
plt.imshow(sarnew, cmap='tab20c')
```

```python
# log2
sarnew = sar_mul_hh_mag.copy() + 1
sarnew = np.log2(sarnew)
plt.imsave(dir.data.base / "image_data" / "log2.png", sarnew, cmap='gray')
plt.imshow(sarnew, cmap='tab20c')
```
